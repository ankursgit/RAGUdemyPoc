### 3. TEXT LOADER - READ SINGLE FILE

```python
from langchain_community.document_loaders import TextLoader

### LOAD SINGLE FILE
loader = TextLoader("data/text_files/python-intro.txt", encoding="UTF-8");
documents = loader.load();

print(type(documents));
print(documents)
```

### 4. TEXT LOADER - MULTIPLE FILES

```python
from langchain_community.document_loaders import DirectoryLoader
from langchain_community.document_loaders import TextLoader

dir_loader = DirectoryLoader(
    "data/text_files",
    glob="**/*.txt",
    loader_cls= TextLoader,
    loader_kwargs={"encoding": "utf-8"},
    show_progress=True
);
documents = dir_loader.load();


print(f"\n Total documents loaded: {len(documents)}\n")
for i, doc in enumerate(documents, start=1):
    # doc.metadata usually contains the source file path
    file_name = doc.metadata.get("source", "Unknown file")
    content_length = len(doc.page_content)
    print(f"{i}. File: {file_name} | Length: {content_length} characters")

```

# TEXT SPLITTER

```python
from langchain_text_splitters import (
    CharacterTextSplitter,
    RecursiveCharacterTextSplitter,
    TokenTextSplitter
)

print(documents)
```

### 1. Character Text Splitter

```python
python_text = documents[1].page_content;
print(f"The Actual Text:\n {python_text}")

print("-----------------------CHARACTER TEXT SPLITTER----------------------");
char_splitter = CharacterTextSplitter(
    separator="",
    chunk_size=200,
    chunk_overlap=20,
    length_function=len
)
chunks = char_splitter.split_text(python_text);
print(f"Total chunks created: {len(chunks)}")
for i, chunk in enumerate(chunks, start=1):
    print(f"Chunk {i}: Length = {len(chunk)} characters")
    #print(f"Content:\n{chunk}\n")
    #print("-" * 80)

print("\n\n-----------------------RECURSIVE CHARACTER TEXT SPLITTER----------------------");
rec_char_splitter = RecursiveCharacterTextSplitter(
    separators=["\n\n", "\n", " ", ""], # follow these splitter in order
    chunk_size=200,
    chunk_overlap=20,
    length_function=len
)
rec_chunks = rec_char_splitter.split_text(python_text);
print(f"Total chunks created: {len(rec_chunks)}\n")
for i, chunk in enumerate(rec_chunks, start=1):
    print(f"Chunk {i}: Length = {len(chunk)} characters")
    #print(f"Content:\n{chunk}\n")
    #print("-" * 80)



print("\n\n-----------------------TOKEN CHARACTER TEXT SPLITTER----------------------");
tok_splitter = TokenTextSplitter(
    chunk_size=200,
    chunk_overlap=20,
)
token_chunks = tok_splitter.split_text(python_text);
print(f"Total chunks created: {len(token_chunks)}\n")
for i, chunk in enumerate(token_chunks, start=1):
    print(f"Chunk {i}: Length = {len(chunk)} characters")
    #print(f"Content:\n{chunk}\n")
    #print("-" * 80)

```
