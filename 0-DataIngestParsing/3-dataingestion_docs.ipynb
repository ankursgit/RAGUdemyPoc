# Word Document Processing

```python
from langchain_community.document_loaders import (
    Docx2txtLoader,
    UnstructuredWordDocumentLoader
)
```

## Method-1: Using Docx2txtLoader

```python
try:
    docx_loader = Docx2txtLoader("data/docs_files/sample-file.docx")
    docs=docx_loader.load()
    print(f"✅ Loaded {len(docs)} document(s)")
    print(f"Length of document: {len(docs[0].page_content)} characters")
    print(f"Metadata: {docs[0].metadata}")

except Exception as e:
    print(f"Error: {e}")
```

## Method-1: Using UnstructuredDoc

```python
try:
    unstructured_loader=UnstructuredWordDocumentLoader("data/docs_files/sample-file.docx",mode="elements")
    unstructured_docs=unstructured_loader.load()
    
    print((unstructured_docs[0]))

    print(f"✅ Loaded {len(unstructured_docs)} elements")
    for i, doc in enumerate(unstructured_docs):
        print(f"\nElement {i+1}:")
        print(f"Type: {doc.metadata.get('category', 'unknown')}")
        print(f"Content: {doc.page_content[:100]}...")


except Exception as e:
    print(e)
```
